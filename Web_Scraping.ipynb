```python
import pandas as pd
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

#Question 1 - Extracting Tesla Stock Data Using yfinance - 2 Points

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period = "max")
tesla_data.reset_index(inplace = True)
tesla_data.head()
```

#Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points

```python
url = " https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data = requests.get(url).text
soup = BeautifulSoup(data, "html.parser")

read_html_data = pd.read_html(str(soup))
tesla_revenue = read_html_data[0]
tesla_revenue.tail()
```

###Then we have to clean the revenue column of "," and "$" characters.

```python
tesla_revenue['Revenue'] = tesla_revenue['Tesla Annual Revenue (Millions of US $).1'].apply(lambda x: x.replace(',', '').replace('$', ''))
tesla_revenue = tesla_revenue.drop('Tesla Annual Revenue (Millions of US $).1', axis = 1)
```

 ### After that, we rename the first column and parse the data type into String for the visualization part

```python
tesla_revenue = tesla_revenue.rename(columns={'Tesla Annual Revenue (Millions of US $)': 'Date'})
tesla_revenue['Date'] = tesla_revenue['Date'].astype(str)
tesla_revenue
```

#Question 3 - Extracting GameStop Stock Data Using yfinance - 2 Points

```python
game_stop = yf.Ticker("GME")
game_stop_data = game_stop.history(period = "max")
game_stop_data.reset_index(inplace = True)
game_stop_data.head()
```

#Question 4 - Extracting GameStop Revenue Data Using Webscraping - 1 Points


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
data  = requests.get(url).text
soup = BeautifulSoup(data, "html.parser")
read_html_pandas_data = pd.read_html(str(soup))
game_stop_revenue= read_html_pandas_data[0]
game_stop_revenue.tail()
```

###Then we repeat the cleaning process to clean the revenue column of "," and "$" characters.

```python
game_stop_revenue['Revenue'] = game_stop_revenue['GameStop Annual Revenue (Millions of US $).1'].apply(lambda x: x.replace(',', '').replace('$', ''))
game_stop_revenue = game_stop_revenue.drop('GameStop Annual Revenue (Millions of US $).1', axis = 1)
```

 ### After that, we rename the first column and parse the data type into String for the visualization part

```python
game_stop_revenue = game_stop_revenue.rename(columns={'GameStop Annual Revenue (Millions of US $)': 'Date'})
game_stop_revenue['Date'] = game_stop_revenue['Date'].astype(str)
game_stop_revenue
```

#First of all we have to define the function to display the dashboards

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=0.3)

    # Filter lines for plotting 2021 and earlier years
    # stock_data_specific = stock_data[stock_data.Date <= '2021--06-14']
    # revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data.Date, infer_datetime_format=True), y=stock_data.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data.Date, infer_datetime_format=True), y=revenue_data.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    fig.show()
```

#Question 5 - Tesla Stock and Revenue Dashboard - 2 Points

In the chart we will see the graph of Tesla's stock prices and earnings, but earnings have not been updated since 2021  

```python
make_graph(tesla_data, tesla_revenue, 'Tesla')
```
